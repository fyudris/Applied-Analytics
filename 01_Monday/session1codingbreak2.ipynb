Load all packages you need.

```python
import pandas as pd
import matplotlib.pyplot as plt
import sweetviz as sv
import numpy as np
```

Read the data.

```python
# Load the data
file_path = '../data/diabetes.csv'
df = pd.read_csv(file_path)

# Display the first few rows
df.head
```

Create an overview of the dataframe (variable names and types; number of observations). What is the target variable? 

```python
# Display the DataFrame info
df.info()

# Display the number of observations
print("\n\nNumber of observations: ")
print(len(df))

# Display the target variable
print("\n\nTarget variable: ")
print ("Outcome")
```

Create a histogram of the variable "BloodPressure".

```python
# Create a histogram of the "BloodPressure" variable
plt.figure(figsize=(10, 6))
plt.hist(df['BloodPressure'], bins=20, edgecolor='k', alpha=0.7)
plt.title('Histogram of BloodPressure')
plt.xlabel('BloodPressure')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Create a scatterplot of the variables "BloodPressure" and "BMI"

```python
# Create a scatterplot of the variables "BloodPressure" and "BMI"
plt.figure(figsize=(10, 6))
plt.scatter(df['BloodPressure'], df['BMI'], alpha=0.7)
plt.title('Scatterplot of BloodPressure vs. BMI')
plt.xlabel('BloodPressure')
plt.ylabel('BMI')
plt.grid(True)
plt.show()
```

Run sweetviz on the dataset and inspect it.

```python
# Generate the Sweetviz report
report = sv.analyze(df)

# Display the report
report.show_html('diabetes_report.html')

# Display the report
report.show_html('diabetes_report.html')
```
